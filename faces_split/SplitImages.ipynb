```python
import os

directories = []
current_dir = os.getcwd()
names_of_people = set()

for element in os.scandir(current_dir):
    if element.is_dir():
        if not "py" in element.name and element.name != 'valid' and element.name != 'train' and element.name != 'test':
            directories.append(current_dir + '\\' + element.name)
        
for directory in directories:
    i = 1
    for image_name in os.listdir(directory):
        parts = directory.split('\\')
        class_name = parts[len(parts)-1]
        names_of_people.add(class_name)
        new_image_name = directory + '\\' + f'Face_{class_name}_{i}.jpg'
        os.rename(directory + '\\' + image_name, new_image_name)
        i = i+1
        
print(names_of_people)
```

```python
import os

directories = []
current_dir = os.getcwd()

#init train/valid/test dirs
test_dir = current_dir + '\\test'
os.mkdir(test_dir)
valid_dir = current_dir + '\\valid'
os.mkdir(valid_dir)
train_dir = current_dir + '\\train'
os.mkdir(train_dir)

images_directories = []
for name in names_of_people:
    images_directories.append(current_dir + '\\' + name)

print(images_directories)
```

```python
import os
import random


current_dir = os.getcwd()


#for each directory
for person_images_dir in images_directories:
    
    n_images = len([entry for entry in os.listdir(person_images_dir) if os.path.isfile(os.path.join(person_images_dir, entry))])
    
    train_max = int(n_images*0.8)
    valid_max = int((n_images - train_max) * 0.8)
    test_max = int(n_images - train_max - valid_max)

    #init split counters
    train_counter = 0
    valid_counter = 0
    test_counter = 0
    
    person_name = person_images_dir.split('_')[-1].split('\\')[-1]
    
    os.mkdir(test_dir + '\\' + person_name)
    os.mkdir(valid_dir + '\\' + person_name)
    os.mkdir(train_dir + '\\' + person_name)
    
    print(f"person {person_name} tot {n_images} train {train_max} valid {valid_max} test {test_max} ")
            
    #for each element
    for element in os.scandir(person_images_dir):
        moved = False
        while not moved:
            #random from 1 to 3
            image_usage_type = random.randrange(1,4)
            if image_usage_type == 1 and train_counter < train_max:
                source = person_images_dir + '\\' + element.name
                destination = train_dir + '\\' + person_name + '\\' + element.name
                print(destination)
                os.system(f'copy {source} {destination}')
                train_counter = train_counter + 1
                moved = True
            if image_usage_type == 2 and valid_counter < valid_max:
                source = person_images_dir + '\\' + element.name
                destination = valid_dir + '\\' + person_name + '\\' + element.name
                print(destination)
                os.system(f'copy {source} {destination}')
                valid_counter = valid_counter + 1
                moved = True
            if image_usage_type == 3 and test_counter < test_max:
                source = person_images_dir + '\\' + element.name
                destination = test_dir + '\\' + person_name + '\\' + element.name
                print(destination)
                os.system(f'copy {source} {destination}')
                test_counter = test_counter + 1
                moved = True
```

```python
import json

with open(current_dir + '\\cat_to_name.json', 'r') as f:
    cat_to_name = json.load(f)
    
print(cat_to_name)
```
